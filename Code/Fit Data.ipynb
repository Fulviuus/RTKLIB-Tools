Here we will generate a secondary data frame that will expand upon the original master containing the true classifications of accuracy using the distances calculated from the stationary point. Below is a big sample where we specify whether appointments within a 30 cm of the ground truth, rendering a spherical boundary of a 30 cm radius. Once the data is classified, the data is randomly shuffled and split into the desired percentages of training, cross validation, and test sets. The data frames are then checked to make sure that time is maintained as an index for the data frame, simply for convenience.

```python
dff = df
X1 = ['Q','ns','sd(m)','ratio']
y1 = 'y'
target_names = ['Outside', 'Inside']
dff[y1] = np.less(np.abs(np.asarray(dff['dist(m)'])),.3)

#n_samples, n_features = dff.shape
#half = int(n_samples / 2)
#dff_train = dff[:half]
#dff_test = dff[half:]
#dff_train = dff[::100]
#dff_test = dff[1::2]

dff_train, cv, dff_test = splitData(dff, 0.8, 0.0, 0.2, seed=42)

dff_train = dff_train.set_index('%_GPST')
dff_test = dff_test.set_index('%_GPST')
```

```python
# Run classifier
svm_rbf = sk.svm.SVC(kernel='rbf', probability=True)
svm_lin = sk.svm.SVC(kernel='linear', probability=True)
```

```python
svm_rbf.fit(dff_train[X1], dff_train[y1])
```

```python
#svm_lin.fit(dff_train[X1], dff_train[y1])
#svm_lin = svm_rbf
```

```python
svm_rbf_probas_ = svm_rbf.predict_proba(dff_test[X1])
#svm_lin_probas_ = svm_lin.predict_proba(dff_test[X1])

dff_test['svm_rbf_probas_'] = svm_rbf_probas_[:, 1]
#dff_test['svm_lin_probas_'] = svm_lin_probas_[:, 1]

svm_rbf_y_ = svm_rbf.predict(dff_test[X1])
#svm_lin_y_ = svm_lin.predict(dff_test[X1])

dff_test['svm_rbf_y_'] = np.asarray(svm_rbf_y_)
#dff_test['svm_lin_y_'] = np.asarray(svm_lin_y_)
```

```python
# Compute ROC curve and area the curve
svm_rbf_fpr, svm_rbf_tpr, svm_rbf_thresholds = roc_curve(dff_test['y'], dff_test['svm_rbf_probas_'])
svm_rbf_roc_auc = auc(svm_rbf_fpr, svm_rbf_tpr)

svm_lin_fpr, svm_lin_tpr, svm_lin_thresholds = roc_curve(dff_test['y'], dff_test['svm_lin_probas_'])
svm_lin_roc_auc = auc(svm_lin_fpr, svm_lin_tpr)
```

```python
print('Generating Fig')
fig1 = pyplot.figure(figsize=(12,12), dpi=200)
fig1.suptitle('Receiver Operating Characteristic', fontsize=14, fontweight='bold')
ax = fig1.add_subplot(111)
#ax = fig1.add_subplot(111,aspect='equal')
p = ax.plot(svm_rbf_fpr, svm_rbf_tpr, label='SVM RBF (area = %0.3f)' % (svm_rbf_roc_auc))
p = ax.plot(svm_lin_fpr, svm_lin_tpr, label='SVM LIN (area = %0.3f)' % (svm_lin_roc_auc))
p = ax.plot([0, 1], [0, 1], 'k--')
pl.legend(loc="lower right")
#xlim(distnminlim, distnmaxlim)
#ylim(disteminlim, distemaxlim)
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.grid(True)
```
